```python
import numpy as np
from scipy. integrate import odeint
import matplotlib.pyplot as plt
import math

t0 = 0
x0 = 15
N = 1500

a1 = 0.89
a2 = 0.000015

t = np.arange( t0, 20, 0.1)

def syst(dx, t):
    x = dx
    return (a1 +x*a2)*(N-x)


y = odeint(syst, x0, t)

fig1 = plt.figure(facecolor='white')
plt.plot(t, y, linewidth=2, label="")
plt.xlabel("t")
plt.ylabel("")
plt.grid(True)
plt.legend()
plt.show()
fig1.savefig('03.png', dpi = 600)

a1 = 0.000015
a2 = 0.82

t = np.arange( t0, 0.1, 0.01)

y = odeint(syst, x0, t)
dy = (a1 +y*a2)*(N-y)

fig2 = plt.figure(facecolor='white')
plt.plot(t, y, linewidth=2, label="")
plt.plot(t, dy, linewidth=2, label="")
plt.xlabel("t")
plt.ylabel("")
plt.grid(True)
plt.legend()
plt.show()
fig2.savefig('04.png', dpi = 600)

def a1(t): 
    a1 = math.sin(9*t)
    return a1

def a2(t): 
    a2 = 0.6*math.sin(4*t)
    return a2

t = np.arange( t0, 0.2, 0.01)

def syst2(dx, t):
    x = dx
    return (a1(t) +x*a2(t))*(N-x)

y = odeint(syst2, x0, t)

fig3 = plt.figure(facecolor='white')
plt.plot(t, y, linewidth=2, label="")
plt.xlabel("t")
plt.ylabel("")
plt.grid(True)
plt.legend()
plt.show()
fig3.savefig('05.png', dpi = 600)
```
